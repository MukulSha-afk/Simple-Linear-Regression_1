```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('height-weight.csv')
```

```python
df.head()
```

```python
import seaborn as sns
plt.scatter(df['Weight'],df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
##deivde our data set into idenpendebt and dependent feature
X=df[['Weight']] ## independent feature
y = df['Height'] ## dependent feature
```

```python
## train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X.shape
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
##standardise the dataset train independent feature data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
plt.scatter(X_train,y_train)
plt.show()
```

```python
##train the simple linear rigression model 
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
print("the slope or coeffiencet of weirght is ",regressor.coef_)
print("Intercept ",regressor.intercept_)
```

```python
plt.scatter(X_train,y_train)
plt.show()
plt.plot(X_train,regressor.predict(X_train),'r' )
```

```python
y_pred_test = regressor.predict(X_test)
```

```python
y_pred_test,y_test
```

```python
plt.scatter(X_test,y_test)
plt.show()
plt.plot(X_test,regressor.predict(X_test),'r+' )
```

## performance matix 

## MSE,MAE,RMSE
##R square and adjusted R quare

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(y_test,y_pred_test)
mae = mean_absolute_error(y_test,y_pred_test)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred_test)
```
